```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt

# Display settings for better visualization
plt.style.use('ggplot')     # Alternative: 'classic', 'dark_background', 'bmh'
pd.options.display.float_format = '{:.6f}'.format
```

```python
class StockDataFetcher:
    def __init__(self, ticker: str, start_date: str, end_date: str):
        """
        Initializes the data fetcher for a given stock ticker.
        :param ticker: Stock symbol (e.g., 'AAPL')
        :param start_date: Start date in 'YYYY-MM-DD' format
        :param end_date: End date in 'YYYY-MM-DD' format
        """
        self.ticker = ticker
        self.start_date = start_date
        self.end_date = end_date
        self.data = None
    
    def fetch_data(self):
        """
        Fetches historical stock data from Yahoo Finance.
        """
        self.data = yf.download(self.ticker, start=self.start_date, end=self.end_date)
        if self.data.empty:
            raise ValueError(f"No data found for {self.ticker}")
        self.data.reset_index(inplace=True)
        print(f"[INFO] Data fetched for {self.ticker}. Rows: {len(self.data)}")
        return self.data
    
    def plot_stock_data(self):
        """
        Plots the historical closing price of the stock.
        """
        if self.data is not None:
            plt.figure(figsize=(10, 5))
            plt.plot(self.data['Date'], self.data['Close'], label='Close Price')
            plt.title(f'{self.ticker} Closing Price')
            plt.xlabel('Date')
            plt.ylabel('Price (USD)')
            plt.legend()
            plt.show()
        else:
            print("[WARNING] No data to plot. Fetch data first.")
```

```python
class TechnicalIndicators:
    def __init__(self, df: pd.DataFrame):
        """
        Initializes the technical indicators generator.
        :param df: DataFrame containing historical stock data
        """
        self.df = df
    
    def add_moving_averages(self):
        """
        Adds Short, Medium, and Long Moving Averages.
        """
        self.df['MA20'] = self.df['Close'].rolling(window=20).mean()
        self.df['MA50'] = self.df['Close'].rolling(window=50).mean()
        self.df['MA200'] = self.df['Close'].rolling(window=200).mean()
    
    def add_rsi(self, period: int = 14):
        """
        Adds Relative Strength Index (RSI).
        :param period: Period for RSI calculation
        """
        delta = self.df['Close'].diff()
        gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
        loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
        rs = gain / loss
        self.df['RSI'] = 100 - (100 / (1 + rs))
    
    def preprocess(self):
        """
        Runs all preprocessing steps and handles NaN values.
        """
        self.add_moving_averages()
        self.add_rsi()
        self.df.bfill(inplace=True)  # Backward fill
        self.df.ffill(inplace=True)  # Forward fill
        print("[INFO] Technical indicators added and NaNs handled.")
        return self.df
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

class DataPreprocessor:
    def __init__(self, df: pd.DataFrame):
        """
        Initializes the data preprocessor for scaling and sequencing.
        :param df: DataFrame containing historical stock data with indicators.
        """
        self.df = df.copy()
        self.scaler = MinMaxScaler()
    
    def normalize_features(self):
        """
        Normalizes numerical features using Min-Max Scaling.
        """
        # ✅ Ensure only numeric columns are kept
        self.df = self.df.select_dtypes(include=[np.number])

        # ✅ Check for NaNs after filtering
        if self.df.isnull().values.any():
            print("[WARNING] NaN values found after dropping non-numeric columns. Filling NaNs.")
            self.df.fillna(method='ffill', inplace=True)
            self.df.fillna(method='bfill', inplace=True)

        features = list(self.df.columns)  # All remaining columns are features
        self.df[features] = self.scaler.fit_transform(self.df[features])
        print("[INFO] Data normalization complete.")
    
    def sequence_data(self, sequence_length: int = 60):
        """
        Transforms the data into sequences for LSTM input.
        :param sequence_length: Number of days to look back for prediction
        """
        X, y = [], []
        # ✅ Clearer indexing
        data = self.df.values
        for i in range(sequence_length, len(data)):
            X.append(data[i-sequence_length:i])
            y.append(data[i][-1])
        
        X, y = np.array(X), np.array(y)
        print(f"[INFO] Sequence data created. Shape of X: {X.shape}, y: {y.shape}")
        return X, y
        preprocessor = DataPreprocessor(stock_data)
        preprocessor.normalize_features()
        X, y = preprocessor.sequence_data(sequence_length=60)
        print(X.shape)   # Should be (samples, 60, features)
        print(y.shape)   # Should be (samples,)
        print(np.isnan(X).any())   # Should be False
        print(np.isnan(y).any())   # Should be False
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Input
from tensorflow.keras.optimizers import Adam
```

```python
import numpy as np

class LSTMModel:
    def __init__(self, input_shape: tuple):
        """
        Initializes the LSTM model.
        :param input_shape: Shape of the input data (e.g., (60, 9) for 60 days and 9 features)
        """
        self.model = Sequential([
            Input(shape=input_shape),             # <--- Correct way to specify input shape
            LSTM(units=50, return_sequences=True),
            Dropout(0.2),
            LSTM(units=50, return_sequences=False),
            Dropout(0.2),
            Dense(units=1)
        ])

        self.model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
        print("[INFO] LSTM model created.")

    def train_model(self, X_train, y_train, epochs=20, batch_size=32, validation_data=None):
        """
        Trains the LSTM model.
        :param X_train: Training feature data
        :param y_train: Training labels (stock prices)
        :param epochs: Number of epochs for training
        :param batch_size: Batch size for training
        :param validation_data: Validation data for evaluating the model during training
        """
        # ✅ Convert data to float32 to avoid dtype errors
        X_train = np.array(X_train).astype(np.float32)
        y_train = np.array(y_train).astype(np.float32)
        
        if validation_data is not None:
            X_val, y_val = validation_data
            X_val = np.array(X_val).astype(np.float32)
            y_val = np.array(y_val).astype(np.float32)
            validation_data = (X_val, y_val)
        
        # ✅ Check for NaN values before training
        if np.isnan(X_train).any() or np.isnan(y_train).any():
            raise ValueError("[ERROR] NaN values detected in training data. Check your preprocessing.")

        history = self.model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=validation_data)
        print(f"[INFO] Model trained for {epochs} epochs.")
        return history

    def evaluate_model(self, X_test, y_test):
        """
        Evaluates the model on test data.
        :param X_test: Test feature data
        :param y_test: Test labels (stock prices)
        """
        X_test = np.array(X_test).astype(np.float32)
        y_test = np.array(y_test).astype(np.float32)
        
        loss = self.model.evaluate(X_test, y_test)
        print(f"[INFO] Model test loss: {loss}")
        return loss

    def predict(self, X_input):
        """
        Predicts stock prices based on the input data.
        :param X_input: Input feature data
        """
        X_input = np.array(X_input).astype(np.float32)
        predictions = self.model.predict(X_input)
        return predictions
```

```python
class TechnicalIndicators:
    def __init__(self, df):
        self.df = df

    def moving_average(self, window=14):
        """Compute the Moving Average (MA)"""
        self.df['MA'] = self.df['Close'].rolling(window=window).mean()
        print("[INFO] Moving Average added.")

    def relative_strength_index(self, window=14):
        """Compute the Relative Strength Index (RSI)"""
        delta = self.df['Close'].diff(1)
        gain = delta.where(delta > 0, 0)
        loss = -delta.where(delta < 0, 0)

        avg_gain = gain.rolling(window=window).mean()
        avg_loss = loss.rolling(window=window).mean()

        rs = avg_gain / avg_loss
        self.df['RSI'] = 100 - (100 / (1 + rs))
        print("[INFO] RSI added.")

    def preprocess(self):
        """Apply all transformations"""
        self.moving_average(window=14)
        self.relative_strength_index(window=14)
        
        # Fill missing values
        self.df.fillna(method='bfill', inplace=True)
        self.df.fillna(method='ffill', inplace=True)
        
        print("[INFO] Preprocessing complete.")
        return self.df
import yfinance as yf

# Fetch the data (you can specify the ticker symbol and period as needed)
df = yf.download('AAPL', start='2018-01-01', end='2025-01-01')
print("[INFO] Data fetched successfully.")


# Initialize the TechnicalIndicators with your dataframe
indicators = TechnicalIndicators(df)
stock_data = indicators.preprocess()

# Display the dataframe to confirm the new features
stock_data.tail()
```
